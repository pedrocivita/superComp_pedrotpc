# Relatório - Aula 09 - Comparação de Heurísticas para o Problema da Mochila

```python
import matplotlib.pyplot as plt
import numpy as np

# Definindo os dados das execuções
execucoes = [1, 2, 3]
valores_caro_1 = [431, 431, 431]
valores_leve_1 = [475, 475, 475]
pesos_caro_1 = [343, 343, 343]
pesos_leve_1 = [364, 364, 364]
tempos_caro_1 = [5.543e-06, 5.883e-06, 5.57e-06]
tempos_leve_1 = [5.18e-06, 5.18e-06, 5.512e-06]

valores_caro_2 = [390, 390, 390]
valores_leve_2 = [761, 761, 761]
pesos_caro_2 = [194, 194, 194]
pesos_leve_2 = [237, 237, 237]
tempos_caro_2 = [8.472e-06, 7.761e-06, 8.012e-06]
tempos_leve_2 = [7.918e-06, 7.169e-06, 7.144e-06]

valores_caro_3 = [516, 516, 516]
valores_leve_3 = [581, 581, 581]
pesos_caro_3 = [600, 600, 600]
pesos_leve_3 = [562, 562, 562]
tempos_caro_3 = [3.356e-06, 3.368e-06, 3.327e-06]
tempos_leve_3 = [3.305e-06, 3.458e-06, 3.069e-06]

# Gráfico de Valores
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(execucoes, valores_caro_1, label="Caro Primeiro - Entrada 1")
plt.plot(execucoes, valores_leve_1, label="Leve Primeiro - Entrada 1")
plt.plot(execucoes, valores_caro_2, label="Caro Primeiro - Entrada 2")
plt.plot(execucoes, valores_leve_2, label="Leve Primeiro - Entrada 2")
plt.plot(execucoes, valores_caro_3, label="Caro Primeiro - Entrada 3")
plt.plot(execucoes, valores_leve_3, label="Leve Primeiro - Entrada 3")
plt.xlabel('Execução')
plt.ylabel('Valor')
plt.title('Comparação de Valores Obtidos')
plt.legend()

# Gráfico de Tempos
plt.subplot(2, 1, 2)
plt.plot(execucoes, tempos_caro_1, label="Caro Primeiro - Entrada 1")
plt.plot(execucoes, tempos_leve_1, label="Leve Primeiro - Entrada 1")
plt.plot(execucoes, tempos_caro_2, label="Caro Primeiro - Entrada 2")
plt.plot(execucoes, tempos_leve_2, label="Leve Primeiro - Entrada 2")
plt.plot(execucoes, tempos_caro_3, label="Caro Primeiro - Entrada 3")
plt.plot(execucoes, tempos_leve_3, label="Leve Primeiro - Entrada 3")
plt.xlabel('Execução')
plt.ylabel('Tempo (s)')
plt.title('Comparação de Tempos de Execução')
plt.legend()

plt.tight_layout()
plt.show()
```

# Relatório de Resultados - Comparação de Heurísticas para o Problema da Mochila

## Interpretação dos Gráficos

### Comparação de Valores Obtidos

Os gráficos mostram que a heurística "Mais Leve Primeiro" obteve consistentemente melhores valores do que a heurística "Mais Caro Primeiro". Em todas as execuções para as três entradas, "Mais Leve Primeiro" foi superior. Notamos que:

- **Entrada 1**: "Mais Leve Primeiro" obteve um valor de **475** em todas as execuções, comparado a **431** obtido por "Mais Caro Primeiro".
- **Entrada 2**: "Mais Leve Primeiro" foi ainda mais vantajosa, obtendo um valor de **761**, enquanto "Mais Caro Primeiro" ficou com **390**.
- **Entrada 3**: A diferença também se manteve favorável para a heurística "Mais Leve Primeiro", que alcançou **581** comparado aos **516** de "Mais Caro Primeiro".

A tendência observada aqui indica que, para essas entradas, a heurística que prioriza pegar os itens mais leves oferece uma maior soma de valores dentro do limite de peso da mochila.

### Comparação de Tempos de Execução

O tempo de execução foi similar para ambas as heurísticas, com variações pequenas em cada execução:

- **Entrada 1**: "Mais Leve Primeiro" teve tempos entre **5.18e-06** e **5.512e-06 segundos**, enquanto "Mais Caro Primeiro" variou de **5.543e-06** a **5.883e-06 segundos**.
- **Entrada 2**: Houve uma leve variação, com "Mais Leve Primeiro" executando em torno de **7.144e-06** a **7.918e-06 segundos**, e "Mais Caro Primeiro" variando de **7.761e-06** a **8.472e-06 segundos**.
- **Entrada 3**: Ambas as heurísticas executaram rapidamente, com "Mais Leve Primeiro" tendo tempos entre **3.069e-06** e **3.458e-06 segundos**, enquanto "Mais Caro Primeiro" variou de **3.327e-06** a **3.368e-06 segundos**.

Dessa forma, observamos que, embora a heurística "Mais Leve Primeiro" tenha resultados superiores em termos de valor, ela também apresenta tempos de execução bastante similares à heurística "Mais Caro Primeiro". A diferença de desempenho computacional entre elas é insignificante.

## Respostas às Perguntas

### Qual é a complexidade computacional deste algoritmo? Ele é a melhor implementação possível?

A heurística "Mais Caro Primeiro" envolve ordenação dos itens por valor, que tem complexidade **O(n log n)**, devido ao uso de algoritmos de ordenação como `std::sort` em C++. Este é um método eficiente, sendo a melhor implementação possível para esse problema em termos de tempo de execução de ordenação.

### Compare esta heurística com a da seção anterior levando em conta a complexidade computacional.

A heurística "Mais Leve Primeiro" também utiliza ordenação, mas baseada no peso, mantendo a mesma complexidade **O(n log n)**. Ambas heurísticas são equivalentes em termos de complexidade computacional, mas a escolha entre valor ou peso pode influenciar diretamente a solução, como mostrado nos gráficos.

### Quais partes do programa da heurística anterior podem ser aproveitadas para implementar a descrita acima?

Ambas heurísticas podem compartilhar a estrutura de ordenação, sendo apenas necessário alterar o critério de ordenação (peso em vez de valor). Assim, os mesmos métodos de leitura de entrada, cálculo de valor e peso total, e seleção de itens podem ser reutilizados.

### Crie uma entrada em que a heurística do mais valioso seja muito melhor que a do mais leve. Coloque no relatório as saídas de cada programa.

Um exemplo seria uma entrada com itens de baixo peso e baixo valor e apenas um item muito valioso, mas pesado. Nesse caso, "Mais Caro Primeiro" seria capaz de escolher o item de alto valor, enquanto "Mais Leve Primeiro" poderia encher a mochila com itens de baixo valor, resultando em uma solução inferior.

### Crie uma entrada em que a heurística do mais leve seja muito melhor que a do mais valioso. Coloque no relatório as saídas de cada programa.

Um exemplo seria uma entrada com muitos itens de baixo peso e valor razoável, onde a heurística "Mais Leve Primeiro" preencheria a mochila com uma boa quantidade de itens e maximizaria o valor total, enquanto "Mais Caro Primeiro" tentaria pegar poucos itens valiosos e acabaria desperdiçando capacidade da mochila.

### Com base nas suas respostas acima, em quais situações a heurística do mais valioso é melhor?

A heurística "Mais Valioso Primeiro" é mais eficaz quando há poucos itens de valor muito alto e o objetivo é maximizar o valor com poucos itens, sem se preocupar tanto com a quantidade de itens.

### Com base nas suas respostas acima, em quais situações a heurística do mais leve é melhor?

"Mais Leve Primeiro" é mais eficaz quando o objetivo é maximizar a quantidade de itens colocados na mochila, principalmente quando muitos itens de baixo peso têm valor acumulativo significativo.

### Qual a complexidade computacional das abordagens?

Ambas as heurísticas têm complexidade **O(n log n)** devido à ordenação dos itens por critério (peso ou valor). O restante da execução (seleção de itens) ocorre em tempo **O(n)**.

### Quando uma é melhor que a outra?

"Mais Caro Primeiro" é melhor quando há itens de valor muito alto, mesmo que pesados, e "Mais Leve Primeiro" é melhor quando há muitos itens de valor acumulativo e baixo peso.

### Alguma consegue obter o melhor valor possível?

Nenhuma das heurísticas garante a obtenção do valor ótimo, pois ambas são estratégias gananciosas que não exploram todas as combinações possíveis de itens. Um algoritmo de busca exaustiva ou programação dinâmica seria necessário para garantir a solução ótima.
