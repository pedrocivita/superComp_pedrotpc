```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Dados para os gráficos

# Entradas
entradas = ['Entrada_1', 'Entrada_2', 'Entrada_3', 'Entrada_4']

# Tempos de Execução (em segundos)
tempos_exaustiva = [1.40465, 89.4821, 0, 9.4271]  # Timeout no caso da Entrada_3
tempos_embaralhada = [
    (1.6116e-05 + 1.5616e-05 + 1.5625e-05 + 1.552e-05 + 1.54e-05) / 5,
    (1.6542e-05 + 1.6126e-05 + 1.6392e-05 + 1.6202e-05 + 1.5453e-05) / 5,
    (1.6023e-05 + 1.6261e-05 + 1.8447e-05 + 1.774e-05 + 1.91e-05) / 5,
    (1.6454e-05 + 1.6855e-05 + 1.4967e-05 + 1.519e-05 + 1.7514e-05) / 5
]
tempos_probabilistica = [
    (1.9347e-05 + 2.0082e-05 + 1.942e-05 + 1.9018e-05 + 2.1321e-05) / 5,
    (2.1466e-05 + 2.4098e-05 + 2.582e-05 + 2.1137e-05 + 2.2079e-05) / 5,
    (2.0213e-05 + 2.0853e-05 + 2.0173e-05 + 1.994e-05 + 2.0998e-05) / 5,
    (1.796e-05 + 1.7404e-05 + 1.7461e-05 + 1.7749e-05 + 1.8832e-05) / 5
]

# Valores alcançados (em dinheiros)
valores_exaustiva = [601, 870, 0, 648]
valores_embaralhada = [
    (190 + 326 + 252 + 202 + 217) / 5,
    (207 + 147 + 97 + 214 + 128) / 5,
    (3320 + 3320 + 3320 + 3320 + 3320) / 5,
    (502 + 445 + 553 + 431 + 438) / 5
]
valores_probabilistica = [
    (275 + 308 + 235 + 327 + 280) / 5,
    (236 + 260 + 275 + 243 + 202) / 5,
    (1746 + 1517 + 2188 + 1706 + 2123) / 5,
    (427 + 381 + 440 + 353 + 440) / 5
]

# Pesos ocupados (em Kg)
pesos_exaustiva = [373, 239, 0, 600]
pesos_embaralhada = [217, 214, 3320, 502]
pesos_probabilistica = [280, 260, 2188, 440]
```

```python
# Definir a posição das barras
bar_width = 0.2
index = np.arange(len(entradas))
```

```python
# Gráfico de Tempo de Execução
plt.figure(figsize=(12, 6))
plt.bar(index, tempos_exaustiva, bar_width, label='Busca Exaustiva')
plt.bar(index + bar_width, tempos_embaralhada, bar_width, label='Heurística Embaralhada')
plt.bar(index + 2 * bar_width, tempos_probabilistica, bar_width, label='Heurística Probabilística')
plt.xlabel('Entradas')
plt.ylabel('Tempo de Execução (segundos)')
plt.title('Comparação de Tempo de Execução')
plt.xticks(index + bar_width, entradas)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gráfico de Valores Alcançados
plt.figure(figsize=(12, 6))
plt.bar(index, valores_exaustiva, bar_width, label='Busca Exaustiva')
plt.bar(index + bar_width, valores_embaralhada, bar_width, label='Heurística Embaralhada')
plt.bar(index + 2 * bar_width, valores_probabilistica, bar_width, label='Heurística Probabilística')
plt.xlabel('Entradas')
plt.ylabel('Valor Alcançado (dinheiros)')
plt.title('Comparação de Valores Alcançados')
plt.xticks(index + bar_width, entradas)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gráfico de Pesos Ocupados
plt.figure(figsize=(12, 6))
plt.bar(index, pesos_exaustiva, bar_width, label='Busca Exaustiva')
plt.bar(index + bar_width, pesos_embaralhada, bar_width, label='Heurística Embaralhada')
plt.bar(index + 2 * bar_width, pesos_probabilistica, bar_width, label='Heurística Probabilística')
plt.xlabel('Entradas')
plt.ylabel('Peso Ocupado (Kg)')
plt.title('Comparação de Pesos Ocupados')
plt.xticks(index + bar_width, entradas)
plt.legend()
plt.grid(True)
plt.show()
```
