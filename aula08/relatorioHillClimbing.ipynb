```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = {
    "Execution": range(1, 11),
    "Heuristic_Aleatoria_Valor": [409, 309, 238, 292, 286, 409, 309, 238, 292, 286],
    "Heuristic_Aleatoria_Tempo": [2e-06, 2e-06, 2e-06, 2e-06, 1e-06, 2e-06, 2e-06, 2e-06, 2e-06, 1e-06],
    "Mochila_Cheia_Valor": [286, 286, 286, 286, 286, 286, 286, 286, 286, 286],
    "Mochila_Cheia_Tempo": [1e-06, 0, 1e-06, 0, 0, 0, 1e-06, 0, 0, 1e-06],
    "Substituicao_Valor": [24154]*10,
    "Substituicao_Tempo": [2.7e-05, 2.5e-05, 2.4e-05, 2.3e-05, 2.3e-05, 2.3e-05, 2.3e-05, 2.3e-05, 2.3e-05, 2.3e-05],
    "Hill_Climbing_Valor": [286]*10,
    "Hill_Climbing_Tempo": [1e-06]*10
}
```

```python
# Criar um DataFrame
df = pd.DataFrame(data)
```

```python
# Plotando os gráficos
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Gráfico dos valores obtidos
ax1.plot(df['Execution'], df['Heuristic_Aleatoria_Valor'], label='Heuristic Aleatoria')
ax1.plot(df['Execution'], df['Mochila_Cheia_Valor'], label='Mochila Cheia')
ax1.plot(df['Execution'], df['Substituicao_Valor'], label='Substituição de Objeto')
ax1.plot(df['Execution'], df['Hill_Climbing_Valor'], label='Hill Climbing')

ax1.set_title('Valores obtidos nas Execuções')
ax1.set_xlabel('Execução')
ax1.set_ylabel('Valor')
ax1.legend()

# Gráfico dos tempos de execução
ax2.plot(df['Execution'], df['Heuristic_Aleatoria_Tempo'], label='Heuristic Aleatoria')
ax2.plot(df['Execution'], df['Mochila_Cheia_Tempo'], label='Mochila Cheia')
ax2.plot(df['Execution'], df['Substituicao_Tempo'], label='Substituição de Objeto')
ax2.plot(df['Execution'], df['Hill_Climbing_Tempo'], label='Hill Climbing')

ax2.set_title('Tempos de Execução')
ax2.set_xlabel('Execução')
ax2.set_ylabel('Tempo (s)')
ax2.legend()

plt.tight_layout()
plt.show()
```

# Análise dos Dados

## Qualidade das Soluções (Gráfico Superior):
- **Substituição de Objeto** obteve consistentemente o valor mais alto em todas as execuções, mostrando claramente uma superioridade sobre as outras estratégias.
- Isso confirma que a Substituição de Objeto explora mais as trocas entre itens e consegue otimizar melhor a solução, refletindo em valores significativamente mais altos.
- **Hill Climbing** e as outras abordagens, como **Mochila Cheia** e **Heurística Aleatória**, retornaram valores muito inferiores em comparação à Substituição de Objeto.
- Essas estratégias, especialmente Hill Climbing, não conseguiram se aproximar do valor ótimo, provavelmente por ficarem presas em ótimos locais.

## Tempos de Execução (Gráfico Inferior):
- **Substituição de Objeto** apresenta um tempo de execução significativamente maior em comparação às outras estratégias, conforme mostrado pela linha verde constante no gráfico.
- Isso reflete o fato de que a abordagem realiza diversas iterações e trocas de itens, resultando em uma solução melhor, mas com um custo de tempo mais elevado.
- **Hill Climbing**, **Mochila Cheia**, e **Heurística Aleatória** apresentaram tempos de execução muito baixos e similares, demonstrando que essas abordagens são muito rápidas.
- Entretanto, essa eficiência de tempo não se reflete em uma melhoria significativa na qualidade das soluções obtidas.

## Conclusão:
- **Substituição de Objeto** é a estratégia que gera as melhores soluções em termos de valor total obtido, mas apresenta um custo de tempo maior. Isso a torna ideal quando a qualidade da solução é mais importante que o tempo de execução.
- **Hill Climbing**, **Mochila Cheia**, e **Heurística Aleatória** são muito rápidas em termos de tempo de execução, mas entregam soluções com valores inferiores. Hill Climbing, embora seja uma estratégia de otimização local, não conseguiu escapar dos máximos locais em nenhum dos casos.

### Considerações:
A escolha da estratégia depende do equilíbrio desejado entre **qualidade da solução** e **tempo de execução**. Para soluções mais rápidas, sem a preocupação de obter o valor máximo possível, Hill Climbing e as heurísticas aleatórias são viáveis. Já para soluções mais otimizadas, mesmo que o tempo seja maior, a **Substituição de Objeto** é a estratégia mais indicada.
