```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = {
    "Execution": range(1, 11),
    "Heuristic_Aleatoria_Valor": [409, 309, 238, 292, 286, 409, 309, 238, 292, 286],
    "Heuristic_Aleatoria_Tempo": [2e-06, 2e-06, 2e-06, 2e-06, 1e-06, 2e-06, 2e-06, 2e-06, 2e-06, 1e-06],
    "Mochila_Cheia_Valor": [286, 286, 286, 286, 286, 286, 286, 286, 286, 286],
    "Mochila_Cheia_Tempo": [1e-06, 0, 1e-06, 0, 0, 0, 1e-06, 0, 0, 1e-06],
    "Substituicao_Valor": [24154]*10,
    "Substituicao_Tempo": [2.7e-05, 2.5e-05, 2.4e-05, 2.3e-05, 2.3e-05, 2.3e-05, 2.3e-05, 2.3e-05, 2.3e-05, 2.3e-05],
    "Hill_Climbing_Valor": [286]*10,
    "Hill_Climbing_Tempo": [1e-06]*10
}
```

```python
# Criar um DataFrame
df = pd.DataFrame(data)
```

```python
# Plotando os grficos
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Grfico dos valores obtidos
ax1.plot(df['Execution'], df['Heuristic_Aleatoria_Valor'], label='Heuristic Aleatoria')
ax1.plot(df['Execution'], df['Mochila_Cheia_Valor'], label='Mochila Cheia')
ax1.plot(df['Execution'], df['Substituicao_Valor'], label='Substituio de Objeto')
ax1.plot(df['Execution'], df['Hill_Climbing_Valor'], label='Hill Climbing')

ax1.set_title('Valores obtidos nas Execues')
ax1.set_xlabel('Execuo')
ax1.set_ylabel('Valor')
ax1.legend()

# Grfico dos tempos de execuo
ax2.plot(df['Execution'], df['Heuristic_Aleatoria_Tempo'], label='Heuristic Aleatoria')
ax2.plot(df['Execution'], df['Mochila_Cheia_Tempo'], label='Mochila Cheia')
ax2.plot(df['Execution'], df['Substituicao_Tempo'], label='Substituio de Objeto')
ax2.plot(df['Execution'], df['Hill_Climbing_Tempo'], label='Hill Climbing')

ax2.set_title('Tempos de Execuo')
ax2.set_xlabel('Execuo')
ax2.set_ylabel('Tempo (s)')
ax2.legend()

plt.tight_layout()
plt.show()
```

# Anlise dos Dados

## Qualidade das Solues (Grfico Superior):
- **Substituio de Objeto** obteve consistentemente o valor mais alto em todas as execues, mostrando claramente uma superioridade sobre as outras estratgias.
- Isso confirma que a Substituio de Objeto explora mais as trocas entre itens e consegue otimizar melhor a soluo, refletindo em valores significativamente mais altos.
- **Hill Climbing** e as outras abordagens, como **Mochila Cheia** e **Heurstica Aleatria**, retornaram valores muito inferiores em comparao  Substituio de Objeto.
- Essas estratgias, especialmente Hill Climbing, no conseguiram se aproximar do valor timo, provavelmente por ficarem presas em timos locais.

## Tempos de Execuo (Grfico Inferior):
- **Substituio de Objeto** apresenta um tempo de execuo significativamente maior em comparao s outras estratgias, conforme mostrado pela linha verde constante no grfico.
- Isso reflete o fato de que a abordagem realiza diversas iteraes e trocas de itens, resultando em uma soluo melhor, mas com um custo de tempo mais elevado.
- **Hill Climbing**, **Mochila Cheia**, e **Heurstica Aleatria** apresentaram tempos de execuo muito baixos e similares, demonstrando que essas abordagens so muito rpidas.
- Entretanto, essa eficincia de tempo no se reflete em uma melhoria significativa na qualidade das solues obtidas.

## Concluso:
- **Substituio de Objeto**  a estratgia que gera as melhores solues em termos de valor total obtido, mas apresenta um custo de tempo maior. Isso a torna ideal quando a qualidade da soluo  mais importante que o tempo de execuo.
- **Hill Climbing**, **Mochila Cheia**, e **Heurstica Aleatria** so muito rpidas em termos de tempo de execuo, mas entregam solues com valores inferiores. Hill Climbing, embora seja uma estratgia de otimizao local, no conseguiu escapar dos mximos locais em nenhum dos casos.

### Consideraes:
A escolha da estratgia depende do equilbrio desejado entre **qualidade da soluo** e **tempo de execuo**. Para solues mais rpidas, sem a preocupao de obter o valor mximo possvel, Hill Climbing e as heursticas aleatrias so viveis.
